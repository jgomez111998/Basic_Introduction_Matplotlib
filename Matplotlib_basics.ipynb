# GRÁFICOS CON MATPLOTLIB

```python
import matplotlib.pyplot as plt
import numpy as np
```

## Gráfico de lineas

```python
x= np.arange(2016,2022)
y= [45,46,47,48,50,51]
```

```python
plt.figure(figsize=(10, 6)) #fija el tamaño de la figura

plt.plot(x,y,marker='o',linestyle='--',color='green', label='Pais1')

#si queremos dibujar otra linea en el gráfico
y2= [40,48,42,44,45,46]
plt.plot(x,y2, marker='o',linestyle='--',color='blue',label='Pais2')

plt.xlabel('Año')
plt.ylabel('Población')
plt.title('Población por año')
plt.legend(loc=4) #muestra los labels

#plt.yticks([45,48,51]) toma esos valores del eje y
#plt.savefig('ejemplo.png') guarda la imagen en la ruta de trabajo
```

# Gráfico de barras

```python
x = ['Argentina','Colombia','Peru']
y = [40,50,33]

plt.bar(x,y) #podemos escribir ejes como en el gráfico de lineas y titulos, cambiar color etc
```

# Gráfico de sectores

```python
plt.pie(y,labels= x) #importante escribir labels = a la variable categórica
```

# Histogramas

```python
import random
random.seed(2)
edad = random.sample(range(100), 60) #60 numeros aleatorios del 0 al 100
bins=[10,30,50,70,90]

plt.hist(edad,bins,edgecolor='black')
```

# Diagrama de caja y bigote o boxplot 

```python
edad.append(250)  #metemos un dato no real, fuera de lo común atípico

plt.boxplot(edad)

plt.show() #nos elimina los comentarios adicionales encima del gráfico
```

# Subplots

```python
#creamos los valores del eje x e y para dibujar el sen(x^2) entre 0 y 2pi
x = np.linspace(0, 2 * np.pi, 400) 
y = np.sin(x ** 2)
```

```python
fig, (ax1, ax2) = plt.subplots(1, 2, figsize=(15,6)) #le decimos que dibuje dos gráficos separados, 1 fila y dos columnas
fig.suptitle('Sen x^2 y -Sen x^2 ') #da nombre general al gráfico
ax1.plot(x,y)
ax2.plot(x,-y)
```
